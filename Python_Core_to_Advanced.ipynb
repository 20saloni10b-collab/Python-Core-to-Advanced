```python
# ============================================================
# MODULE 5 : LOOPS
# ============================================================

# Loops are used to execute a block of code repeatedly.


# ============================================================
# WHILE LOOP
# ============================================================

# Print numbers from 1 to 5

count = 1

while count <= 5:
    print(count)
    count += 1


# Print even numbers from 2 to 10

number = 2

while number <= 10:
    print(number)
    number += 2


# Countdown Timer

timer = 5

while timer > 0:
    print(timer)
    timer -= 1

print("Time Up")


# Banking Example

balance = 1000

while balance <= 5000:
    print("Balance:", balance)
    balance += 1000


# Student Attendance Example

attendance_days = 1

while attendance_days <= 5:
    print("Day", attendance_days)
    attendance_days += 1


# ============================================================
# FOR LOOP
# ============================================================

# Print numbers from 0 to 4

for i in range(5):
    print(i)


# Print numbers from 1 to 5

for i in range(1, 6):
    print(i)


# Iterate through a string

for character in "Python":
    print(character)


# Iterate through a list

subjects = ["Python", "AI", "ML"]

for subject in subjects:
    print(subject)


# Student Names

students = ["Rahul", "Aman", "Riya"]

for student in students:
    print(student)


# Shopping Cart Example

cart_items = ["Laptop", "Mouse", "Keyboard"]

for item in cart_items:
    print(item)


# ============================================================
# NESTED LOOPS
# ============================================================

# Number Pattern

for i in range(3):
    for j in range(3):
        print(i, j)


# Star Pattern

for row in range(3):
    for col in range(3):
        print("*", end=" ")
    print()


# Multiplication Table Pattern

for i in range(1, 4):
    for j in range(1, 4):
        print(i * j, end=" ")
    print()


# Classroom Example

for classroom in range(1, 4):
    for student in range(1, 4):
        print("Class", classroom, "Student", student)


# ============================================================
# INFINITE LOOPS
# ============================================================

# Infinite loops continue forever until stopped manually.

# Example 1

# while True:
#     print("Running...")


# Example 2

# count = 1
#
# while count > 0:
#     print(count)


# Example 3

# for i in iter(int, 1):
#     print(i)


# Note:
# Infinite loops are commonly used in servers,
# games, chat applications and monitoring systems.


# ============================================================
# BREAK STATEMENT
# ============================================================

# break exits the loop immediately.

for i in range(10):

    if i == 5:
        break

    print(i)


# Student Example

students = ["Rahul", "Aman", "Riya", "Rohit"]

for student in students:

    if student == "Riya":
        break

    print(student)


# Banking Example

transactions = [1000, 2000, -5000, 3000]

for amount in transactions:

    if amount < 0:
        print("Suspicious Transaction Found")
        break

    print(amount)


# ============================================================
# CONTINUE STATEMENT
# ============================================================

# continue skips the current iteration.

for i in range(10):

    if i == 5:
        continue

    print(i)


# Print Odd Numbers

for number in range(1, 11):

    if number % 2 == 0:
        continue

    print(number)


# Student Example

marks = [90, 35, 85, 20, 78]

for mark in marks:

    if mark < 40:
        continue

    print(mark)


# ============================================================
# PASS STATEMENT
# ============================================================

# pass is a placeholder statement.

for i in range(5):

    if i == 3:
        pass

    print(i)


# Empty Function

def future_feature():
    pass


# Empty Class

class Student:
    pass


print("Program Continues")


# ============================================================
# RANGE FUNCTION
# ============================================================

# range(stop)

for i in range(5):
    print(i)


# range(start, stop)

for i in range(1, 6):
    print(i)


# range(start, stop, step)

for i in range(1, 11, 2):
    print(i)


# Reverse Counting

for i in range(10, 0, -1):
    print(i)


# Multiplication Table

for i in range(1, 11):
    print("5 x", i, "=", 5 * i)


# Year Example

for year in range(2020, 2026):
    print(year)


# ============================================================
# ENUMERATE FUNCTION
# ============================================================

# enumerate() provides both index and value.

subjects = ["Python", "AI", "ML"]

for index, subject in enumerate(subjects):
    print(index, subject)


# Starting index from 1

students = ["Rahul", "Aman", "Riya"]

for roll_no, student in enumerate(students, start=1):
    print(roll_no, student)


# Shopping Example

products = ["Laptop", "Mouse", "Keyboard"]

for item_no, product in enumerate(products, start=101):
    print(item_no, product)


# City Example

cities = ["Mohali", "Chandigarh", "Delhi"]

for index, city in enumerate(cities):
    print(index, city)


# AI Example

models = ["Linear Regression", "Random Forest", "CNN"]

for model_no, model in enumerate(models, start=1):
    print(model_no, model)
```

#**Module 11: Advanced Data Structures**

```python
# ============================================================
# MODULE 11 : ADVANCED DATA STRUCTURES
# ============================================================

# Python provides several built-in and specialized data
# structures for efficient data storage and manipulation.

# In this module, we will cover:
# 1. Arrays
# 2. Stack
# 3. Queue
# 4. Deque
# 5. Heap Basics


# ============================================================
# ARRAYS
# ============================================================

# Arrays store elements of the same data type.

from array import array

numbers = array('i', [10, 20, 30, 40, 50])

print(numbers)

# Access Elements

print(numbers[0])
print(numbers[2])

# Add Element

numbers.append(60)

print(numbers)

# Insert Element

numbers.insert(1, 15)

print(numbers)

# Remove Element

numbers.remove(30)

print(numbers)

# Iterate Through Array

for number in numbers:
    print(number)

# Student Marks Example

marks = array('i', [85, 90, 78, 92])

for mark in marks:
    print(mark)

# Banking Example

transactions = array('i', [1000, 2500, 5000])

print(transactions)

# AI Example

epochs = array('i', [10, 20, 30])

print(epochs)


# ============================================================
# STACK
# ============================================================

# Stack follows LIFO
# Last In First Out

stack = []

# Push Operation

stack.append("Python")
stack.append("AI")
stack.append("ML")

print(stack)

# Pop Operation

removed_item = stack.pop()

print("Removed:", removed_item)

print(stack)

# Peek Operation

print("Top Element:", stack[-1])

# Check Stack Size

print("Size:", len(stack))

# Browser History Example

history = []

history.append("Google")
history.append("YouTube")
history.append("ChatGPT")

print(history)

history.pop()

print(history)

# Book Stack Example

books = []

books.append("Book 1")
books.append("Book 2")
books.append("Book 3")

print(books)


# ============================================================
# QUEUE
# ============================================================

# Queue follows FIFO
# First In First Out

from collections import deque

queue = deque()

# Enqueue

queue.append("Customer 1")
queue.append("Customer 2")
queue.append("Customer 3")

print(queue)

# Dequeue

served_customer = queue.popleft()

print("Served:", served_customer)

print(queue)

# Queue Size

print(len(queue))

# Ticket Counter Example

tickets = deque()

tickets.append("Person A")
tickets.append("Person B")
tickets.append("Person C")

print(tickets)

tickets.popleft()

print(tickets)

# Print Queue

documents = deque()

documents.append("Doc1")
documents.append("Doc2")
documents.append("Doc3")

print(documents)


# ============================================================
# DEQUE
# ============================================================

# Deque stands for Double Ended Queue.
# Elements can be added or removed from both ends.

from collections import deque

dq = deque()

# Insert at Right

dq.append(10)
dq.append(20)

print(dq)

# Insert at Left

dq.appendleft(5)

print(dq)

# Remove from Right

dq.pop()

print(dq)

# Remove from Left

dq.popleft()

print(dq)

# Student Line Example

students = deque()

students.append("Rahul")
students.append("Aman")

students.appendleft("Monitor")

print(students)

# Undo Feature Example

actions = deque(["Open", "Edit", "Save"])

print(actions)


# ============================================================
# HEAP BASICS
# ============================================================

# Heap is a specialized tree-based data structure.
# Python provides min-heap using heapq module.

import heapq

numbers = [30, 10, 50, 20, 40]

heapq.heapify(numbers)

print(numbers)

# Smallest Element

print(heapq.heappop(numbers))

print(numbers)

# Insert Element

heapq.heappush(numbers, 5)

print(numbers)

# Peek Minimum Element

print(numbers[0])

# Student Marks Example

marks = [90, 70, 85, 60, 95]

heapq.heapify(marks)

print(marks)

print(heapq.heappop(marks))

# Banking Example

transactions = [5000, 1000, 7000, 2000]

heapq.heapify(transactions)

print(transactions)

# AI Example

loss_values = [0.45, 0.30, 0.55, 0.20]

heapq.heapify(loss_values)

print(loss_values)

print(heapq.heappop(loss_values))
```

```python
# ============================================================
# MODULE 14 : ADVANCED FUNCTIONS
# ============================================================

# Advanced functions help write concise, reusable,
# and efficient code.


# ============================================================
# LAMBDA FUNCTIONS
# ============================================================

# Lambda functions are small anonymous functions.
# Syntax:
# lambda arguments : expression

square = lambda x: x * x

print(square(5))


# Addition

add = lambda a, b: a + b

print(add(10, 20))


# Multiplication

multiply = lambda a, b: a * b

print(multiply(5, 6))


# Even or Odd

check_even = lambda x: x % 2 == 0

print(check_even(10))
print(check_even(7))


# Student Example

grade = lambda marks: "Pass" if marks >= 40 else "Fail"

print(grade(75))
print(grade(30))


# Banking Example

interest = lambda principal, rate: principal * rate / 100

print(interest(10000, 5))


# Product Example

discount_price = lambda price, discount: price - discount

print(discount_price(50000, 5000))


# AI Example

accuracy = lambda correct, total: (correct / total) * 100

print(accuracy(95, 100))


# ============================================================
# RECURSIVE FUNCTIONS
# ============================================================

# A recursive function calls itself.

# Factorial Example

def factorial(n):

    if n == 1:
        return 1

    return n * factorial(n - 1)

print(factorial(5))


# Sum of Natural Numbers

def natural_sum(n):

    if n == 0:
        return 0

    return n + natural_sum(n - 1)

print(natural_sum(10))


# Countdown

def countdown(n):

    if n == 0:
        return

    print(n)

    countdown(n - 1)

countdown(5)


# Power Calculation

def power(base, exponent):

    if exponent == 0:
        return 1

    return base * power(base, exponent - 1)

print(power(2, 4))


# Fibonacci Series

def fibonacci(n):

    if n <= 1:
        return n

    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(6))


# ============================================================
# NESTED FUNCTIONS
# ============================================================

# Functions defined inside another function.

def outer():

    print("Outer Function")

    def inner():
        print("Inner Function")

    inner()

outer()


# Student Example

def student():

    name = "Rahul"

    def display():
        print(name)

    display()

student()


# Banking Example

def account():

    balance = 50000

    def show_balance():
        print(balance)

    show_balance()

account()


# Product Example

def product():

    price = 65000

    def display_price():
        print(price)

    display_price()

product()


# ============================================================
# HIGHER ORDER FUNCTIONS
# ============================================================

# Higher-order functions can:
# 1. Accept functions as arguments
# 2. Return functions

def greet(name):
    return f"Hello {name}"

def execute(func, value):
    print(func(value))

execute(greet, "Rahul")


# Calculator Example

def square(number):
    return number * number

def process(function, value):
    print(function(value))

process(square, 5)


# Banking Example

def bonus(amount):
    return amount + 5000

process(bonus, 25000)


# Using map()

numbers = [1, 2, 3, 4, 5]

result = list(map(lambda x: x * 2, numbers))

print(result)


# Using filter()

numbers = [1, 2, 3, 4, 5, 6]

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

print(even_numbers)


# ============================================================
# ANONYMOUS FUNCTIONS
# ============================================================

# Lambda functions are anonymous functions
# because they do not have a formal name.

square = lambda x: x ** 2

print(square(4))


# Sorting Example

students = [
    ("Rahul", 90),
    ("Aman", 85),
    ("Riya", 95)
]

students.sort(key=lambda student: student[1])

print(students)


# Product Example

products = [
    ("Laptop", 65000),
    ("Mouse", 500),
    ("Keyboard", 1500)
]

products.sort(key=lambda item: item[1])

print(products)


# Banking Example

accounts = [
    ("Rahul", 25000),
    ("Aman", 50000),
    ("Rohit", 15000)
]

accounts.sort(key=lambda account: account[1])

print(accounts)


# ============================================================
# CLOSURES
# ============================================================

# A closure remembers values from its enclosing scope.

def outer_function(message):

    def inner_function():
        print(message)

    return inner_function

closure = outer_function("Welcome to Python")

closure()


# Student Example

def student_name(name):

    def display():
        print("Student:", name)

    return display

student = student_name("Rahul")

student()


# Banking Example

def account_balance(balance):

    def display():
        print("Balance:", balance)

    return display

account = account_balance(50000)

account()


# Product Example

def product_price(price):

    def display():
        print("Price:", price)

    return display

product = product_price(65000)

product()


# AI Example

def model_accuracy(accuracy):

    def display():
        print("Accuracy:", accuracy)

    return display

model = model_accuracy(98.5)

model()


# Closure with Calculation

def multiplier(x):

    def multiply(y):
        return x * y

    return multiply

double = multiplier(2)

print(double(10))

triple = multiplier(3)

print(triple(10))
```

```python
# ============================================================
# MODULE 23 : STATIC & CLASS METHODS
# ============================================================

# In Python, methods inside a class can be:
#
# 1. Instance Methods
# 2. Static Methods
# 3. Class Methods
#
# Important Keywords:
# - self
# - cls
# - @staticmethod
# - @classmethod


# ============================================================
# SELF KEYWORD
# ============================================================

# self refers to the current object of a class.
# It is used to access instance variables and methods.

class Student:

    def __init__(self, name, course):

        self.name = name
        self.course = course

    def display(self):

        print("Name:", self.name)
        print("Course:", self.course)

student1 = Student("Rahul", "Python")

student1.display()


# Example 2

class Employee:

    def __init__(self, name, salary):

        self.name = name
        self.salary = salary

    def show_details(self):

        print(self.name)
        print(self.salary)

employee = Employee("Aman", 50000)

employee.show_details()


# Banking Example

class BankAccount:

    def __init__(self, holder, balance):

        self.holder = holder
        self.balance = balance

    def show_balance(self):

        print("Holder:", self.holder)
        print("Balance:", self.balance)

account = BankAccount("Rohit", 75000)

account.show_balance()


# Product Example

class Product:

    def __init__(self, name, price):

        self.name = name
        self.price = price

    def product_info(self):

        print(self.name)
        print(self.price)

laptop = Product("Laptop", 65000)

laptop.product_info()


# AI Example

class Model:

    def __init__(self, model_name, accuracy):

        self.model_name = model_name
        self.accuracy = accuracy

    def display(self):

        print(self.model_name)
        print(self.accuracy)

model = Model("CNN", 98.5)

model.display()


# ============================================================
# STATIC METHODS
# ============================================================

# Static methods belong to the class.
# They do not use self or cls.
# They are defined using @staticmethod.

class Calculator:

    @staticmethod
    def add(a, b):

        return a + b

print(Calculator.add(10, 20))


# Example 2

class MathOperations:

    @staticmethod
    def square(number):

        return number * number

print(MathOperations.square(5))


# Student Example

class Student:

    @staticmethod
    def welcome_message():

        print("Welcome Students")

Student.welcome_message()


# Banking Example

class Bank:

    @staticmethod
    def calculate_interest(principal, rate):

        return principal * rate / 100

print(Bank.calculate_interest(10000, 5))


# Product Example

class Product:

    @staticmethod
    def gst(price):

        return price * 0.18

print(Product.gst(50000))


# AI Example

class AIModel:

    @staticmethod
    def calculate_accuracy(correct, total):

        return (correct / total) * 100

print(AIModel.calculate_accuracy(95, 100))


# Utility Example

class Temperature:

    @staticmethod
    def celsius_to_fahrenheit(celsius):

        return (celsius * 9/5) + 32

print(Temperature.celsius_to_fahrenheit(30))


# ============================================================
# CLASS METHODS
# ============================================================

# Class methods work with class variables.
# They use cls instead of self.
# They are defined using @classmethod.

class Student:

    college = "C-DAC Mohali"

    @classmethod
    def display_college(cls):

        print(cls.college)

Student.display_college()


# Example 2

class Employee:

    company = "C-DAC"

    @classmethod
    def company_name(cls):

        print(cls.company)

Employee.company_name()


# Updating Class Variable

class Course:

    duration = 3

    @classmethod
    def update_duration(cls, months):

        cls.duration = months

Course.update_duration(6)

print(Course.duration)


# Student Example

class Student:

    total_students = 0

    def __init__(self):

        Student.total_students += 1

    @classmethod
    def show_count(cls):

        print(cls.total_students)

s1 = Student()
s2 = Student()
s3 = Student()

Student.show_count()


# Banking Example

class Bank:

    bank_name = "State Bank"

    @classmethod
    def display_bank(cls):

        print(cls.bank_name)

Bank.display_bank()


# Product Example

class Product:

    category = "Electronics"

    @classmethod
    def display_category(cls):

        print(cls.category)

Product.display_category()


# AI Example

class Model:

    framework = "PyTorch"

    @classmethod
    def display_framework(cls):

        print(cls.framework)

Model.display_framework()


# ============================================================
# INSTANCE METHOD VS STATIC METHOD VS CLASS METHOD
# ============================================================

class Demo:

    company = "C-DAC"

    def __init__(self, name):

        self.name = name

    # Instance Method
    def instance_method(self):

        print("Instance Method")
        print(self.name)

    # Class Method
    @classmethod
    def class_method(cls):

        print("Class Method")
        print(cls.company)

    # Static Method
    @staticmethod
    def static_method():

        print("Static Method")

obj = Demo("Rahul")

obj.instance_method()

Demo.class_method()

Demo.static_method()


# ============================================================
# REAL WORLD EXAMPLE
# ============================================================

class Employee:

    company_name = "C-DAC Mohali"

    def __init__(self, name, salary):

        self.name = name
        self.salary = salary

    def display_employee(self):

        print(self.name)
        print(self.salary)

    @classmethod
    def display_company(cls):

        print(cls.company_name)

    @staticmethod
    def calculate_bonus(salary):

        return salary * 0.10

employee = Employee("Aman", 50000)

employee.display_employee()

Employee.display_company()

print(Employee.calculate_bonus(50000))
```
